```python
import numpy as np
import os 
import sys
import pandas as pd
```

```python
from keras.models import Sequential,Model
from keras.layers import *
from keras.datasets import mnist
from keras.optimizers import Adam
from keras.layers.advanced_activations import LeakyReLU
from tqdm import tqdm
```

```python
(X_t,Y_t),(X_te,Y_te) = mnist.load_data()
```

```python
X_train = np.reshape(X_t,(60000,28,28,1)).astype('float32')
X_test = np.reshape(X_te,(10000,28,28,1)).astype('float32')
```

```python
#Normalize images between -1 to 1
X_train = (X_train - 127.5)/127.5
```

```python
#Generator model (input : random images )
def generator():
    model = Sequential()
    model.add(Dense(128*7*7,input_dim = 100 , activation = LeakyReLU(0.2)))
    model.add(BatchNormalization())
    model.add(Reshape((7,7,128)))
    model.add(UpSampling2D())
    model.add(Conv2D(64,(5,5), padding = 'same' , activation = LeakyReLU(0.2)))
    model.add(BatchNormalization())
    model.add(UpSampling2D())
    model.add(Conv2D(1,(5,5), padding = 'same' , activation = 'tanh'))
    return model
```

```python
gen_model = generator()
print(gen_model.summary())
```

```python
#Discriminator model ( input : real images + fake images from generator output )
def discriminator():
    model = Sequential()
    model.add(Conv2D(64,(5,5),padding = 'same',input_shape = (28,28,1) , activation = LeakyReLU(0.2) , subsample = (2,2)))
    #model.add(BatchNormalization())
    model.add(Dropout(0.3))
    model.add(Conv2D(128,(5,5),padding = 'same', activation = LeakyReLU(0.2) , subsample = (2,2)))
    #model.add(BatchNormalization())
    model.add(Dropout(0.3))
    model.add(Flatten())
    model.add(Dense(1,activation = 'sigmoid'))
    return model
```

```python
disc_model = discriminator() 
disc_model.summary()
```

```python
disc_model.compile(loss = 'binary_crossentropy' , optimizer = Adam() )
gen_model.compile(loss = 'binary_crossentropy' , optimizer = Adam() )
```

```python
#Build Adverserial Model using gen and disc 
gen_inp = Input(shape = (100,))

gen_out = gen_model(gen_inp)

disc_model.trainable = False

disc_out = disc_model(gen_out)

adv_model = Model(input = gen_inp , output = disc_out)     
```

```python
adv_model.summary()
adv_model.compile(loss = 'binary_crossentropy' , optimizer = Adam() )
```

```python
def train(epochs):
    batch_size = 128
    batch = 400
    for i in range(epochs):
        for j in tqdm(range(batch)):
            #Noise for input to generator
            noise_1 = np.random.rand(batch_size,100)
            #Random images output from generator
            gen_images = gen_model.predict(noise_1 , batch_size = batch_size )
            #Real images from MNIST dataset
            image_batch = X_train[np.random.randint(0, X_train.shape[0], size=batch_size)]
            #Creating input data for discriminator (real images + generator output images)
            disc_inp = np.concatenate([gen_images,image_batch])
            #Labels (1 for real and 0 for fake ) for discriminator training 
            disc_Y = [0]*batch_size + [1]*batch_size
            #Make discriminator model trainable
            disc_model.trainable = True
            #Train discriminator model
            disc_model.train_on_batch(disc_inp,disc_Y)
            #Generate noise for the adversarial network
            noise_2 = np.random.rand(batch_size,100)
            #Freeze Discriminator weights and train generator(of the adversarial model)
            disc_model.trainable = False
            #Labels for adversarial model will always be 1
            y_adv = [1]*batch_size
            #Train Adversarial model
            adv_model.train_on_batch(noise_2,y_adv)
            #Repeat for 'epoch' no. of times
```

```python
#20 epochs 
train(20)
```

```python
#save weights 
gen_model.save_weights('gen_20_scaled_images.h5')
disc_model.save_weights('dis_20_scaled_images.h5')
```

```python
#load weights
gen_model.load_weights('gen_20_scaled_images.h5')
disc_model.load_weights('dis_20_scaled_images.h5')
```

```python
#Display output
import matplotlib.pyplot as plt
def plot_output(text):
    try_input = np.random.rand(50, 100) #Random noise as input
    preds = gen_model.predict(try_input)
    plt.figure(figsize=(20,20))
    for i in range(preds.shape[0]):
        plt.subplot(10, 10, i+1)
        plt.imshow(preds[i, :, :, 0], cmap='gray')
        plt.axis('off')
    
    # tight_layout minimizes the overlap between 2 sub-plots
    plt.tight_layout()
    plt.savefig(text)
```

```python
plot_output('20')
```

```python
#50 total epochs
train(30)
```
